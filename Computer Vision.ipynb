```python
'''
Script for a Naive Bayes Classifier, which attempts to distinguish
between handwritten digits 0,1,...,9. 

The handwritten digits are represented through ASCII art in the
trainingimages and testimages files.

First, the classifier runs through a training set of 5000 handwritten digits
to set up the Maximum a Posteriori (MAP) and Maximum Likelihood (ML) estimations.

Then, the classifier's accuracy is tested through a testing set of 1000 handwritten digits.
After running through the testing set, this script will display a confusion matrix, most 
prototypical handwritten digit for 0-9, log likelihood maps, and log odd ratio maps between 
each pair of digits.
'''
```

```python
#Imports for the Classifier
import numpy as np;
import numpy.linalg as la; 
import matplotlib.pyplot as pt;
import math as math;

#Set the home directory for file access.
%cd "/home/owner/cs440/mp3/part1"
#Open up the training file for handwritten digits.
training_file = open("trainingimages");
training_labels_file = open("traininglabels");

#F contains the training set of 500 handwritten digits in 28x28 pixel images.
#F_identites contains the correct digit label of each handwritten digit in the training set.
F = np.ndarray((500,28,28));
F_identities = np.ndarray((500,1));

#For loop to fill F and F_identities from the trainingimages and traininglabels files.
for img in range(500):
    input_line = training_labels_file.readline();
    F_identities[img] = int(input_line);
    
    for i in range(28):
        input_line = training_file.readline();
        
        for j in range(28):
            if (input_line[j] == '+' or input_line[j] == '#'):
               F[img][i][j] = 1;
            else:
                F[img][i][j] = 0;
#End for loop
```

```python
'''
Training the Classifier by filling out its
MAP and ML estimations.
'''
```

```python

#Constants
V = 2; #Num of values a feature(pixel) can take
k = 1; #A smoothening constant to V. More k = more smoothening
'''
This 4-dimensional array consists of 28x28 pixels that can take 2 values given 10 possible digit classes:
array[i][j][1][d] = 1 => pixel at (i,j) with value 1 (foreground) for digit d
array[i][j][0][d] = 1 => pixel at (i,j) with value 0 (background) for digit d
'''

likelihood_of_F_at_pixel_given_class = np.zeros((28,28,2,10)); 
#num_trainings_per_class contains
#the number of occurences of the 10 digit classes in the 500 training examples.
num_trainings_per_class = np.zeros((10,1)); 

#Get the numerator of the likelihood probabilities
for img in range(500):
    actual_class = (int)(F_identities[img]);
    
    for i in range(28):
        for j in range(28):
            if (F[img][i][j] == 1):
                likelihood_of_F_at_pixel_given_class[i][j][1][actual_class] += 1;
            else:
                likelihood_of_F_at_pixel_given_class[i][j][0][actual_class] += 1;

#Get the denominator of the likelihood probabilities
for img in range(500):
    occurence = (int)(F_identities[img]);
    num_trainings_per_class[occurence] += 1;
    
#Laplacian Smoothening of the likelihood numerators and denominators
likelihood_of_F_at_pixel_given_class += k;
num_trainings_per_class += k*V;

#Divide the likelihood numerator with the likelihood denominator to obtain the likelihood probabilities.
for i in range(28):
    for j in range(28):
        for value in range(2):
            for num_class in range(10):
                likelihood_of_F_at_pixel_given_class[i][j][value][num_class] /= float(num_trainings_per_class[num_class]);

#Find the prior probabilities.
prior_prob_class = num_trainings_per_class/500;
```

```python
#Block to do memory freeing and filestream closing.
F = None;
F_ind = None;
training_file.close();
training_labels_file.close();
```

```python
'''
Testing the Classifier by going through the 
testing set.
'''
```

```python
#Open up the testing file for handwritten digits.
test_file = open("testimages");
#test contains the testing set of 100 handwritten digits in 28x28 pixel images.
test = np.ndarray((100,28,28));

#For loop to fill test from the testimages file.
for img in range(100):
    for i in range(28):
        input_line = test_file.readline();
        for j in range(28):
            if (input_line[j] == '+' or input_line[j] == '#'):
                test[img][i][j] = 1;
            else:
                test[img][i][j] = 0;
#End for loop
```

```python
#Here, we will create the MAP and ML estimations for the testing set.
#This array contains 100 handwritten digits to classify via full-Naive Bayes.
map_classification = np.zeros((100,1));
#This array contains 100 handwritten digits to classify via Naive Bayes without the prior probability.
ml_classification = np.zeros((100,1));

#This array contains 10 testing set images with the highest posterior probability
#to a unique digit class's likelihoods (from the training set).
highest_posterior_prob_per_class = np.zeros((10,28,28));
#This array contains the indices where the aforementioned 10 images are located in the testing set.
highest_posterior_prob_index = np.zeros((10,)); 
#This array contains the max posterior probabilities of each of the 10 digits.
highest_posterior_prob_value = np.zeros((10,));

#We initialize the max posterior probabilities to be in the negatives for the Max Search Loop
#that appears later in this block.
highest_posterior_prob_value -= 1000;


'''
Test the classifier by iterating through all 100 test examples. For each test example,
have the classifier decide what digit a test example is (through MAP and ML estimation), 
and store it in map_classification and ml_classification respectively.
'''
for img in range(100):
    #We will run each testing set picture through 10 digit classes.
    posterior_of_class_given_pixel = np.zeros((10,)); 
    #test_image is the current 28x28 testing set picture
    test_image = test[img]; 
    
    #run through each digit class to test if the current testing set picture belongs there.
    for num_class in range(10):
        '''
        Add log(F_11 | class) + log(F_12 | class) + ... to each element of posterior_of_class_given_pixel.
        
        The reason I'm doing log-adding instead of straight multiplication is because a 0 term in a product
        results in the product becoming 0. As a result, there is a chance all elements in posterior_of_class_given_pixel
        could be 0, making the classifier as ineffective as a random guess.
        '''
        for i in range(28):
            for j in range(28):
                if (test_image[i][j] == 1):
                    posterior_of_class_given_pixel[num_class] += math.log2(likelihood_of_F_at_pixel_given_class[i][j][1][num_class]);
                else:
                    posterior_of_class_given_pixel[num_class] += math.log2(likelihood_of_F_at_pixel_given_class[i][j][0][num_class]);
        
    #Now, for ML_classification, we won't add log(P(class))    
    #ML_classification (ML estimate)
    maxInd = -1;
    maxVal = -1000;
    for i in range(10):
        if (posterior_of_class_given_pixel[i] > maxVal):
            maxInd = i;
            maxVal = posterior_of_class_given_pixel[i];
    
    '''
    Using ML Estimate, the classifier assigns the current test example
    a digit label. A digit label is chosen if its corresponding entry in
    posterior_of_class_given_pixel is the largest entry in the array.
    '''
    ml_classification[img] = maxInd;
        
    #Add log(P(class)) for MAP_classification
    for num_class in range(10):
        posterior_of_class_given_pixel[num_class] += math.log2(prior_prob_class[num_class]);
        
    #MAP_classification (MAP estimate)
    maxInd = -1;
    maxVal = -1000;
    for i in range(10):
        if (posterior_of_class_given_pixel[i] > maxVal):
            maxInd = i;
            maxVal = posterior_of_class_given_pixel[i];
            
        #This unique section of code that's not in the ML classification loop
        #is used to find a testing picture that's closest to a digit's posterior probability.
        if (posterior_of_class_given_pixel[i] > highest_posterior_prob_value[i]):
            highest_posterior_prob_value[i] = posterior_of_class_given_pixel[i];
            highest_posterior_prob_index[i] = img;
            highest_posterior_prob_per_class[i] = test_image;
        
    '''
    Using MAP Estimate, the classifier assigns the current test example
    a digit label. A digit label is chosen if its corresponding entry in
    posterior_of_class_given_pixel is the largest entry in the array.
    '''
    map_classification[img] = maxInd;
#end for loop
```

```python
#close the first testing set filestream.
test_file.close();
```

```python
#Open up the testing file for handwritten digits.
testlabels_file = open("testlabels");
#correct_answers contains the correct label of the elements within the testing set.
correct_answers = np.ndarray((100,1)); 
#number_of_classes counts the occurence of the 10 unique digit classes.
#This will be used for the confusion matrix.
number_of_classes = np.zeros((10,1));

#fill up correct_answers
for i in range(100):
    input_line = testlabels_file.readline();
    correct_answers[i] = int(input_line);
    number_of_classes[int(input_line)] += 1;

#close the other testing set filestream.
testlabels_file.close();
```

```python
#Now we find the testing set accuracy for MAP and ML estimation...
accuracy_of_ml = 0;
accuracy_of_map = 0;
#And the confusion matrices for MAP and ML estimation.
ml_confusion_matrix = np.zeros((10,10)); #Row = actual label, col = predicted/classified label
map_confusion_matrix = np.zeros((10,10)); #Row = actual label, col = predicted/classified label

#Go through all of the classifier's labels and true labels and create the MAP/ML
#confusion matrices and MAP/ML accuracies.
for i in range(100):
    if (ml_classification[i] == correct_answers[i]):
        accuracy_of_ml += 1;
    ml_confusion_matrix[ int(correct_answers[i]) ][ int(ml_classification[i]) ] += 1;
    
    if (map_classification[i] == correct_answers[i]):
        accuracy_of_map += 1;
    map_confusion_matrix[ int(correct_answers[i]) ][ int(map_classification[i]) ] += 1;

#Dividing rows in the confusion matrix with the number of the classes in the testing set.
for i in range(10):
    ml_confusion_matrix[i] /= number_of_classes[i];
    map_confusion_matrix[i] /= number_of_classes[i];
```

```python
#Print the MAP/ML accuracies.
print("ML derived the right digit " + repr(accuracy_of_ml) + " percent of the time.")
print("MAP derived the right digit " + repr(accuracy_of_map) + " percent of the time.")
np.set_printoptions(precision=3, suppress=True);
```

```python
'''
This code prints the ML Estimate Confusion Matrix
in a more viewable format.
'''
columns = ('guess 0', 'guess 1', 'guess 2', 'guess 3', 
        'guess 4', 'guess 5', 'guess 6', 'guess 7', 'guess 8', 'guess 9');
rows = ('        ', 'actual 0', 'actual 1', 'actual 2', 'actual 3', 'actual 4', 'actual 5', 
        'actual 6', 'actual 7', 'actual 8', 'actual 9');

table_to_write = [];

#table_to_write is 11x1 with this for loop.
#Creates row axis
for i,j in enumerate(rows):
    table_to_write.append([]);
    table_to_write[i].append(j); #i = index, j = rows[i]
#Creates column axis
for i,j in enumerate(columns):
    table_to_write[0].append(j); #i = index, j = columns[i]
#Puts ML Values in.
#table_to_write is 11x11 with this
for row in range(1,11):
    for col in range(0,10):
        #Rounded to fifth decimal
        table_to_write[row].append('%.4f' % ml_confusion_matrix[row-1][col]);
        

print("Confusion Matrix for ML (in percentage form)\n")

# from https://docs.python.org/2/tutorial/inputoutput.html
for row in range(0, 11):
        print(repr(table_to_write[row][0]).rjust(10),
              repr(table_to_write[row][1]).rjust(10),
              repr(table_to_write[row][2]).rjust(10),
              repr(table_to_write[row][3]).rjust(10),
              repr(table_to_write[row][4]).rjust(10),
              repr(table_to_write[row][5]).rjust(10),
              repr(table_to_write[row][6]).rjust(10),
              repr(table_to_write[row][7]).rjust(10),
              repr(table_to_write[row][8]).rjust(10),
              repr(table_to_write[row][9]).rjust(10),
              repr(table_to_write[row][10]).rjust(10));

print("\n");

'''
This code prints the MAP Estimate Confusion Matrix
in a more viewable format.
'''
table_to_write = [];

#table_to_write is 11x1 with this.
#Creates row axis
for i,j in enumerate(rows):
    table_to_write.append([]);
    table_to_write[i].append(j); #i = index, j = rows[i]
#Creates column axis
for i,j in enumerate(columns):
    table_to_write[0].append(j); #i = index, j = columns[i]
#Puts ML Values in.
#table_to_write is 11x11 with this
for row in range(1,11):
    for col in range(0,10):
        table_to_write[row].append('%.4f' % map_confusion_matrix[row-1][col]);
        #Rounded to fifth decimal

print("Confusion Matrix for MAP (in percentage form)\n");

# from https://docs.python.org/2/tutorial/inputoutput.html
for row in range(0, 11):
        print(repr(table_to_write[row][0]).rjust(10),
              repr(table_to_write[row][1]).rjust(10),
              repr(table_to_write[row][2]).rjust(10),
              repr(table_to_write[row][3]).rjust(10),
              repr(table_to_write[row][4]).rjust(10),
              repr(table_to_write[row][5]).rjust(10),
              repr(table_to_write[row][6]).rjust(10),
              repr(table_to_write[row][7]).rjust(10),
              repr(table_to_write[row][8]).rjust(10),
              repr(table_to_write[row][9]).rjust(10),
              repr(table_to_write[row][10]).rjust(10));

print("\n");
print("Observe the diagonals to see the classification rates (in percentage form) of each number");
```

```python
'''
This code block prints the images in the testing set with the greatest
posterior probability to each digit's likelihoods (that were developed in
the training set)
'''
print("Test examples with highest posterior probability\n");
for i in range(10):
    elem = highest_posterior_prob_per_class[i];
    pt.figure();
    pt.title("The most prototypical digit " + repr(i) + '\n'
             "is on index " + repr(highest_posterior_prob_index[i]) + " of the 100 tests" + '\n'
             "with posterior probability " + repr(2**(highest_posterior_prob_value[i])) + '\n');
    pt.imshow(elem);
```

```python
'''
This code block prints classifier mistakes that are interesting to look at
and show the downsides of Naive Bayes Classification.
'''
print("Interesting Incorrect Classifications");
#This array contains 4 28x28 images of misclassified handwritten digits.
incorrect = np.zeros((4,28,28));
#This boolean array is used as a medium to find those 4 interesting images.
toSkip = np.zeros((4,));
'''
Finding the four specific interesting incorrect classifications.

Based on the MAP Confusion Matrix, 9 is confused for 4, 3 is confused for 5,
5 is confused for 6, and 9 is confused for 7.
'''
for i in range(100):
    guess = int(map_classification[i]);
    answer = int(correct_answers[i]);
    if (guess != answer):
        if (guess == 9 and answer == 4 and toSkip[0] == 0):
            incorrect[0] = test[i];
            toSkip[0] = 1;
        elif (guess == 3 and answer == 5 and toSkip[1] == 0):
            incorrect[1] = test[i];
            toSkip[1] = 1;
        elif (guess == 5 and answer == 6 and toSkip[2] == 0):
            incorrect[2] = test[i];
            toSkip[2] = 1;
        elif (guess == 9 and answer == 7 and toSkip[3] == 0):
            incorrect[3] = test[i];
            toSkip[3] = 1;


elem = incorrect[0];
pt.figure();
pt.title("AI guessed 9");
pt.imshow(elem);
elem = incorrect[1];
pt.figure();
pt.title("AI guessed 3");
pt.imshow(elem);
elem = incorrect[2];
pt.figure();
pt.title("AI guessed 5");
pt.imshow(elem);
elem = incorrect[3];
pt.figure();
pt.title("AI guessed 9");
pt.imshow(elem);
```

```python
'''
For the log odds ratio map, we will only use MAP classification's results.
'''
```

```python
'''
This code block prints out the 4 most-confused pairs of digits
in the testing set classification.
'''
#confusion_pairs is a 4x2 matrix containing the 4 most-confused pairs of digits.
confusion_pairs = np.zeros((4,2)); #[i][0] = actual digit label. [i][1] = guess of digit label.
confusion_pairs -= 1; #Initialize the matrix to not hold a possible digit (like 0,1,2,..)


#Find the top 4 most confused pairs of digits.
for rank in range(4):
    maxVal = -1000;
    
    for row in range(10):
        for col in range(10):
            if (row != col): #We don't count correct guesses.
            
                inPairAlready = False;
                #This part checks if the (row,col) is already in confused_pairs
                for i in range(4):
                    if (confusion_pairs[i][0] == row and confusion_pairs[i][1] == col):
                        inPairAlready = True;
                
                if (map_confusion_matrix[row][col] > maxVal and inPairAlready == False):
                    maxVal = map_confusion_matrix[row][col];
                    confusion_pairs[rank][0] = row;
                    confusion_pairs[rank][1] = col;
#End for loop

print(confusion_pairs);
```

```python
'''
This code block takes the most-confused pairs of digits
and does two things:
1) Prints the log likelihood heat map of each digit (hot color = more likely,
   cold color = less likely)
2) Prints the log odds ratio heat map of a guessed digit label over the 
   actual digit label. (hot color = more likely, cold color = less likely)
'''
for rank in range(4):
    guess_index = int(confusion_pairs[rank][1]);
    actual_index = int(confusion_pairs[rank][0]);
    guess_map = np.zeros((28,28));
    actual_map = np.zeros((28,28));
    odd_map = np.zeros((28,28));
    for row in range(28):
        for col in range(28):
            guess_map[row][col] = math.log2(likelihood_of_F_at_pixel_given_class[row][col][1][guess_index]);
            actual_map[row][col] = math.log2(likelihood_of_F_at_pixel_given_class[row][col][1][actual_index]);
            #Division is log difference
            odd_map[row][col] = guess_map[row][col] - actual_map[row][col];
    pt.figure();
    pt.title("log likelihood map of " + repr(guess_index));
    pt.imshow(guess_map);
    pt.figure();
    pt.title("log likelihood map of " + repr(actual_index));
    pt.imshow(actual_map);
    pt.figure();
    pt.title("log odds ratio for " + repr(guess_index) + " over " + repr(actual_index));
    pt.imshow(odd_map);
```
